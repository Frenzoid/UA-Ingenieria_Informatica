
## Visualización en celda de texto colab de una imagen compartida en google drive

Este es un enlace que se genera al **obtener enlace para compartir** archivo (público para leer) en google drive:

https://drive.google.com/file/d/1mhc5YGKcskwzW4LUKgn8SJ4wZZC58Kxa/view?usp=sharing

el id de la imagen es: 1mhc5YGKcskwzW4LUKgn8SJ4wZZC58Kxa

Para visualizar una imagen compartida de google dirve en una celda de texto colab, se utiliza esta sintaxis:

&lt;img src= "https://drive.google.com/uc?export=view&id=id de la imagen">
