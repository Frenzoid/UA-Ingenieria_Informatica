###La sentencia If###

En el ejemplo, la sentencia de control comienza con un if, seguido de la expresión a comprobar, seguida de ':'

Debajo hay un bloque de código, con cuatro espacios de sangría, que se ejecuta si la comprobación (x > 0,0) es verdadera:

```python
x = -10.0  # Initial x value

if x > 0.0:  
    print('Initial x is greater than zero')
    x -= 20.0
elif x < 0.0:  
    print('Initial x is less than zero')
    x += 21.0
else: 
    print('Initial x is not less than zero and not greater than zero, therefore it must be zero')
    x *= 2.5

# Print new x value
print("New x value:", x)
```

###Blucle for###

```python
for n in range(4):
    print("----")
    print(n, n**2)
```

Lo anterior ejecuta 4 veces, sobre los enteros 0, 1, 2 y 3, por defecto empieza desde cero (ver https://docs.python.org/3/library/stdtypes.html#range para la documentación de range).
```python
for i in range(-2, 3):
    print(i)
```

El bucle comienza en -2, pero no incluye el 3. Si queremos pasar por tres en lugar de por uno:

```python
for n in range(0, 10, 3):
    print(n)
```

###break, continue y pass###

A veces queremos salir de un bucle for o while. Tal vez en un bucle for podemos comprobar si algo es verdadero, y luego salir del bucle prematuramente, por ejemplo

```python
for x in range(10):
    print(x)
    if x == 5:
        print("Time to break out")
        break
```

```python
N = 50  # Check numbers up 50 for primes (excludes 50)

# Loop over all numbers from 2 to 50 (excluding 50)
for n in range(2, N):

    # Assume that n is prime
    n_is_prime = True
    # Check if n can be divided by m, where m ranges from 2 to n (excluding n)
    for m in range(2, n):
         if n % m == 0:   # Esto es cierto si el resto de n/m es igual a cero
                          # Hemos comprobado que n es divisible por m, por lo que no puede ser un número primo. 
                          # No es necesario comprobar más valores de m, así que establece n_es_primo = Falso y
                          # salimos del bucle 'm'.
            n_is_prime = False
            break

    #  If n is prime, print to screen        
    if n_is_prime:
        print(n)
```

```python
for j in range(20):
    if j % 4 == 0:  # Check remained of j/4
        continue  # jump to next iteration over j
    print("Number is not divisible by 4:", j)
```

```python
for x in range(20):
    if x < 5:
        #  lo haré en el futuro
        pass  
    elif x < 9:
        print(x*x)
    else:
        print(x)
```

Bucles infinitos: causa y protección
Un error común, especialmente cuando se utilizan sentencias while, es el bucle infinito. Esto ocurre cuando se introduce un bucle pero nunca termina (sale). Los bucles infinitos pueden hacer que un sistema no responda, a veces requiriendo un apagado para restaurar la función.

Es una buena práctica, especialmente cuando se está aprendiendo, añadir protecciones contra los bucles infinitos. Por ejemplo,

```python
x = 0.0

counter = 0
while x < 0.05:

    # Guard against infinite loop
    counter += 1
    if counter > 2000:
        print("Loop count exceeded 2000. Exiting")
        break
```

Ejemplo: representación inexacta.
Es trivial que

𝑥=11𝑥-10𝑥 

Si 𝑥=0,1 , podemos escribir

𝑥=11𝑥-1

El error de redondeo para el caso de 0,1 es pequeño, y en muchos casos no supondrá un problema. Sin embargo, a veces los errores de redondeo pueden acumularse y destruir la precisión:
Ahora, partiendo de 𝑥=0,1 evaluamos el lado derecho para obtener un 'nuevo' 𝑥 , y utilizamos este nuevo 𝑥 para evaluar de nuevo el lado derecho. La aritmética es trivial: 𝑥 debe seguir siendo igual a 0,1 . Probamos esto en un programa que repite este proceso 20 veces:

```python
x = 0.1
for i in range(20):
    x = x*11 - 1
    print(x)
```

La solución estalla y se desvía ampliamente de 𝑥=0,1 . Los errores de redondeo se amplifican en cada paso, lo que lleva a una respuesta completamente errónea. La representación informática de 0,1 no es exacta, y cada vez que multiplicamos 0,1 por 11 , aumentamos el error en un factor de 10 aproximadamente (podemos ver arriba que perdemos un dígito de precisión en cada paso). Se puede observar el mismo problema utilizando programas de hojas de cálculo.

Si utilizamos 𝑥=0,5 , que se puede representar exactamente en binario:

```python
x = 0.5
for i in range(20):
    x = x*11 - 5
    print(x)
```

El resultado es exacto en este caso.

Por defecto, Python utiliza 64 bits para almacenar un float. Podemos utilizar el módulo NumPy para crear un float que utilice sólo 32 bits. Probando esto para el caso 𝑥=0.1:

```python
import numpy as np
x = np.float32(0.1)
for i in range(20):
    x = x*11 - 1
    print(x)
```

El error se dispara más rápido en este caso comparado con el de 64 bits - usar 32 bits lleva a una peor aproximación de 0,1 que cuando se usan 64 bits.

Nota: Algunos lenguajes tienen herramientas especiales para realizar aritmética decimal (base 10) (por ejemplo, https://docs.python.org/3/library/decimal.html). Esto permitiría, por ejemplo, representar exactamente 0,1. Sin embargo, el decimal no es la aritmética "natural" de los ordenadores, por lo que cabe esperar que las operaciones en decimal sean mucho más lentas.
